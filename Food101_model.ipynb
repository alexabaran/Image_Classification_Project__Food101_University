Powstała jako udoskonalenie wcześniejszej wersji MobileNet (V1), oferując lepszą dokładność przy mniejszej liczbie parametrów i mniejszych wymaganiach sprzętowych.
Jest wytrenowany na dużym zbiorze obrazów ImageNet i dobrze sprawdza się w zadaniach rozpoznawania obrazów. W projekcie został dostosowany do klasyfikacji 10 potraw.

**ImageNet** to ogromna baza danych zawierająca ponad 14 milionów obrazów w tysiącach kategorii. Służy do trenowania i testowania modeli rozpoznawania obrazów.

Cechy MobileNetV2:
- Bardzo mały rozmiar modelu – idealny do uruchamiania na telefonach, Raspberry Pi itp.
- Szybkość działania – szybki trening i inferencja.
- Inverted Residual Blocks – unikalna struktura bloków z tzw. „butelkowaniem informacji”.
- Zoptymalizowany pod kątem transfer learningu – można łatwo dostosować do własnego zadania (np. klasyfikacja jedzenia).

### A. Prosty CNN

Trzywarstwowa sieć konwolucyjna z warstwami gęstymi i dropoutem, używana do klasyfikacji 10 klas potraw.

```python
model_cnn = nn.Sequential(
    nn.Conv2d(3, 16, kernel_size=3, padding=1),   # wejście RGB - 3 kanały i użycie 16 filtrów konwolucyjnych i rozmiarach 3x3, Padding =1 -nie zmieniamy rozmiaru obrazu
    nn.ReLU(),                                    # Relu - dodaje nieliniowość
    nn.MaxPool2d(2, 2),                           # zmniejsza rozmiar z 224 na 112

    nn.Conv2d(16, 32, kernel_size=3, padding=1),
    nn.ReLU(),
    nn.MaxPool2d(2, 2),

    nn.Conv2d(32, 64, kernel_size=3, padding=1),
    nn.ReLU(),
    nn.MaxPool2d(2, 2),

    nn.Flatten(),
    nn.Linear(64 * 28 * 28, 256),
    nn.ReLU(),
    nn.Dropout(0.5),                              # wyłączmy 50% neuronów (aby uniknąć overfittingu)
    nn.Linear(256, 10)                            # 10 klas wyjściowych
).to(device)
```

```python
print(model_cnn)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model_cnn.parameters(), lr=0.001)
```

```python
# Przyjmujemy 20 epok do obliczeń:
epochs = 20
train_losses = []
train_accuracies = []

# trening modelu
for epoch in range(epochs):
    model_cnn.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)     # dane muszą być przeniesiona na nsz device zgodnie z modelem

        optimizer.zero_grad()           #resetuje gradienty z poprzednich kroków
        outputs = model_cnn(images)
        loss = criterion(outputs, labels)     #obliczamy błedy predykcji
        loss.backward()
        optimizer.step()        # aktualizacja wag na podstawie obliczonych gradientów

        running_loss += loss.item()

        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    avg_loss = running_loss / len(train_loader)
    accuracy = 100 * correct / total
    train_losses.append(avg_loss)
    train_accuracies.append(accuracy)

    print(f"Epoch {epoch+1}/{epochs} — Train Loss: {avg_loss:.4f} — Train Accuracy: {accuracy:.2f}%")
```

### B. MobileNet v2

```python
# Ładujemy już przetrenowany model
model_mobilenet = mobilenet_v2(pretrained=True)
```

```python
# musimy podmienić ostatnią warstwę naszą warstwą dopasowaną do 10 klas
model_mobilenet.classifier[1] = nn.Linear(model_mobilenet.last_channel, 10)

# cały model zawsze musi być na tym samym device (CPU lub GPU)
model_mobilenet = model_mobilenet.to(device)
```

```python
# takie same paramtery jak dla prostego modelu - optimizer Adam i learning rate = 0.001
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model_mobilenet.parameters(), lr=0.001)
```

```python
epochs = 10
train_losses = []
train_accuracies = []

for epoch in range(epochs):
    model_mobilenet.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model_mobilenet(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    avg_loss = running_loss / len(train_loader)
    accuracy = 100 * correct / total
    train_losses.append(avg_loss)
    train_accuracies.append(accuracy)

    print(f"Epoch {epoch+1}/{epochs} — Train Loss: {avg_loss:.4f} — Train Accuracy: {accuracy:.2f}%")
```

Prosty CNN:
- Trening przebiega stopniowo, z rosnącą dokładnością (od 23% do 62%).
- Loss systematycznie spada.
- Tempo uczenia jest wolniejsze (co normalne przy prostych modelach).

MobileNet v2:
- Już na 1. epoce 70% accuracy, czyli znacznie lepiej niż CNN na końcu.
- Do 10. epoki osiąga około 89% accuracy — to bardzo wysoki wynik.

## II. oszacowanie stopnia dopasowania modelu wg dowolnie przyjętych założeń

Stopień dopasowania modelu został oceniony na podstawie jego skuteczności w klasyfikacji obrazów w zbiorze testowym (walidacyjnym). Przyjęliśmy, że podstawową miarą oceny będzie accuracy, czyli procent poprawnie sklasyfikowanych przykładów spośród wszystkich danych wejściowych.

Model został przełączony w tryb ewaluacji (eval()), a obliczenia zostały przeprowadzone z wyłączonym śledzeniem gradientów (torch.no_grad()), co przyspiesza działanie i zapobiega modyfikacjom wag. Dla każdej próbki model wygenerował etykietę klasy, która została porównana z rzeczywistą etykietą z danych walidacyjnych. Na tej podstawie zliczono trafne predykcje oraz całkowitą liczbę przykładów, a wynik accuracy obliczono jako:

Przyjęto podział danych na 80% do treningu i 20% do walidacji, z równomiernym rozkładem klas (stratyfikacja). Dane obejmowały 10 klas potraw, takich jak pizza, sushi, dumplings, itp.

### A. Prosty CNN

```python
# ewaluacja modelu
model_cnn.eval()
correct = 0
total = 0

with torch.no_grad():     #wyłączamy śledzenie gradientów (potrzebne tylko przy trenowaniu)
    for images, labels in val_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model_cnn(images)             # przepuszcza obrazy przez model
        _, predicted = torch.max(outputs, 1)    # przewiduje etykiete
        correct += (predicted == labels).sum().item()   #porównujemy przewidywane z rzeczywistymi i sumujemy
        total += labels.size(0)

accuracy_cnn = 100 * correct / total
print(f"Validation Accuracy: {accuracy_cnn:.2f}%")
```

Wyniki prostego modelu CNN:
Train accuracy (po 20 epokach): 61.11%

Validation accuracy: 55.20%

Model stopniowo uczył się klasyfikować obrazy - dokładność rosła powoli, osiągając nieco powyżej 60% na danych treningowych.

Różnica między treningiem a walidacją jest umiarkowana, co sugeruje przyzwoitą ogólność, ale też ograniczoną moc modelu - być może wynikającą z jego uproszczonej architektury.

### B. MobileNet v2

```python
model_mobilenet.eval()
correct = 0
total = 0

with torch.no_grad():
    for images, labels in val_loader:
        images, labels = images.to(device), labels.to(device)

        outputs = model_mobilenet(images)
        _, predicted = torch.max(outputs, dim=1)  # jawne dim=1 dla czytelności

        correct += (predicted == labels).sum().item()
        total += labels.size(0)

accuracy_mobilenet = 100 * correct / total
print(f"MobileNetv2 Validation Accuracy: {accuracy_mobilenet:.2f}%")
```

Wyniki modelu MobileNetV2 (transfer learning):
Train accuracy (po 10 epokach): 89.38%

Validation accuracy: 86.15%

Model już po pierwszej epoce osiągnął 70% trafności, a jego skuteczność bardzo szybko rosła.

Niewielka różnica między wynikiem na zbiorze treningowym a walidacyjnym świadczy o bardzo dobrej generalizacji.

Dzięki wcześniejszemu wytrenowaniu na dużym zbiorze (ImageNet), model był w stanie szybko przystosować się do klasyfikacji potraw, nawet przy ograniczonym zbiorze danych.

```python
print(f"CNN Validation Accuracy: {accuracy_cnn:.2f}%")
print(f"MobileNet v2 Validation Accuracy: {accuracy_mobilenet:.2f}%")
```

**WNIOSKI:**

- Prosty CNN mimo poprawnych wyników wymaga więcej epok i lepszych danych, aby dorównać zaawansowanym architekturom.

- W kontekście praktycznym - jeśli liczy się wydajność i dokładność, użycie gotowego modelu z transfer learningiem jest zdecydowanie bardziej efektywne.
